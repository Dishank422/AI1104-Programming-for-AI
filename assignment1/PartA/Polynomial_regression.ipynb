```python
#the following code implements the linear regressor y = f(x,theta) = b0 + b1x +...+bmx, N = 10, m = 2
import numpy as np
import matplotlib.pyplot as plt

#Number of training samples = N
N = 10

#Number of parameters = m+1
m = 2

#Generate equispaced samples in the interval [0,2pi] (both 0 and 2*pi excluded)
x = np.linspace(2*np.pi/(N+1),2*np.pi,N,endpoint = False)

#Genrate noise
mean = 0
std = 0.05

#Generate some numbers from the sine function
y = np.sin(x)

#Add noise
y += np.random.normal(mean,std,N)

#Constructing X matrix
X = np.ones(N)
for i in range(m):
    X = np.append(X,x**(i+1))
X = np.reshape(X,(m+1,N))
X = np.transpose(X)

#B = [b0]
#    [b1]
#    [b2]
#As derived in class, B = (X_transpose*X)^(-1)*(X_transpose*y)
B = np.matmul(np.linalg.inv(np.matmul(np.transpose(X),X)),np.matmul(np.transpose(X),y))

#Calculating mean square error while training
y_predicted = np.matmul(B,np.transpose(X))
mean_square_error = (1/N)*np.matmul(y_predicted - y, np.transpose(y_predicted - y))
print("Training error = ")
print(mean_square_error)

#Testing steps

#Number of test samples = T
T = 10
x_test = np.linspace(2*np.pi/(T+1),2*np.pi,T,endpoint = False)

#Genrate noise
mean = 0
std = 0.05

#Generate some numbers from the sine function
y_test = np.sin(x_test)

#Add noise
y_test += np.random.normal(mean,std,T)

#Constructing X_test
X_test = np.ones(T)
for i in range(m):
    X_test = np.append(X_test,x_test**(i+1))
X_test = np.reshape(X_test,(m+1,T))
X_test = np.transpose(X_test)

#Predicting the output
y_predicted = np.matmul(B,np.transpose(X_test))

#Calculating mean square error while testing
mean_square_error = (1/T)*np.matmul(y_predicted - y_test, np.transpose(y_predicted - y_test))
print("Testing error = ")
print(mean_square_error)

#Displaying the results
plt.plot(x_test, y_test,'bx', label = 'ground truth')
plt.plot(x_test, y_predicted,'r', label = 'predicted by regressor')
plt.legend(loc="upper right")
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression model')
plt.show()

#printing observations
print("Observations: The model is underfitting")
```
