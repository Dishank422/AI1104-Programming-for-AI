```python
# Importing packages and data 
import pandas as pd
import random
complete_dataset = pd.read_csv("rating_complete.csv")
```

```python
# Previewing data
complete_dataset.head(50)
```

```python
# Understanding data
rows = complete_dataset.shape[0]
print(rows)
```

```python
# Finding useful variable max_anime_id
max_anime_id = complete_dataset['anime_id'].max()
print(max_anime_id)
```

```python
# Finding useful variable max_user_id
max_user_id = complete_dataset['user_id'].max()
print(max_user_id)
```

```python
# Counting number of views each anime got, number of anime each user has watched and 
# storing start index of each user in complete dataset so that we can easily access any user later on
no_of_anime_watched = []
no_of_views = []
start_index = []

for i in range(max_anime_id+1):
    no_of_views.append(0)
    
for i in range(max_user_id+1):
    no_of_anime_watched.append(0)
    start_index.append(0)

for i in range(rows):
    user_id = complete_dataset['user_id'].iloc[i]
    anime_id = complete_dataset['anime_id'].iloc[i]
    no_of_anime_watched[user_id] += 1
    no_of_views[anime_id] += 1
    if user_id != complete_dataset['user_id'].iloc[i-1]:
        start_index[user_id] = i
```

```python
# We take anime with high views only
# Here, we are counting the number of such anime
# I chose anime which more than 42000 users watched because their number was around 250
# I chose training data of size 5000, so I maintained ratio of users to number of anime as in complete dataset

count = 0
for i in range(max_anime_id+1):
    if no_of_views[i] > 42000:
        count += 1
count
```

```python
# Generating random unique samples, apendding them to lists and generating final DataFrames
# Training sample has 5000 users, testing and validation sets have 1000 each
# Users who have watched more shows than 40 were specifically chose to reduce sparsity
# Therefore, a random sample was generated which was more than 7000 in size.
# There might be instances where a user has watched more than 40 anime but none of them have more than 42000 views
# In such cases, a user will not be added. 
# Thus few more users were chosen added to list than required and later, the extras were manually delete

temp_list1 = []
temp_list2 = []
temp_list3 = []
counter = 0
k = 0

random_sample = random.sample(range(max_user_id), 20000)

while counter < 5007:
    user_id = random_sample[k]

    if no_of_anime_watched[user_id] >= 40 and start_index[user_id+1] != 0:
        counter += 1

        for i in range(start_index[user_id],start_index[user_id+1]):
            anime_id = complete_dataset['anime_id'].iloc[i]
            if no_of_views[anime_id] >= 42000:
                temp_list1.append(user_id)
                temp_list2.append(anime_id)
                temp_list3.append(complete_dataset['rating'].iloc[i])
    k += 1

training_data = pd.DataFrame(list(zip(temp_list1,temp_list2,temp_list3)), columns = ['user_id','anime_id','rating'])

temp_list1.clear()
temp_list2.clear()
temp_list3.clear()

counter = 0

while counter < 1001:
    user_id = random_sample[k]

    if no_of_anime_watched[user_id] >= 40 and start_index[user_id+1] != 0:
        counter += 1
        index = start_index[user_id]

        for i in range(start_index[user_id],start_index[user_id+1]):
            anime_id = complete_dataset['anime_id'].iloc[i]
            if no_of_views[anime_id] >= 42000:
                temp_list1.append(user_id)
                temp_list2.append(anime_id)
                temp_list3.append(complete_dataset['rating'].iloc[i])
    k += 1

validation_data = pd.DataFrame(list(zip(temp_list1,temp_list2,temp_list3)), columns = ['user_id','anime_id','rating'])

temp_list1.clear()
temp_list2.clear()
temp_list3.clear()

counter = 0

while counter < 1001:
    user_id = random_sample[k]

    if no_of_anime_watched[user_id] >= 40 and start_index[user_id+1] != 0:
        counter += 1
        index = start_index[user_id]

        for i in range(start_index[user_id],start_index[user_id+1]):
            anime_id = complete_dataset['anime_id'].iloc[i]
            if no_of_views[anime_id] >= 42000:
                temp_list1.append(user_id)
                temp_list2.append(anime_id)
                temp_list3.append(complete_dataset['rating'].iloc[i])
    k += 1

testing_data = pd.DataFrame(list(zip(temp_list1,temp_list2,temp_list3)), columns = ['user_id','anime_id','rating'])
```

```python
# Priting number of users in each data set
print(training_data['user_id'].unique().size,validation_data['user_id'].unique().size,testing_data['user_id'].unique().size)
```

```python
# Deleting extras manually and verifying using above cell
training_data.drop(index=training_data.index[-1:], 
        axis=0, 
        inplace=True)
training_data.tail()
```

```python
# Saving training, validation and testing data
training_data.to_csv("training_data.csv",index=False)
validation_data.to_csv("validation_data.csv",index=False)
testing_data.to_csv("testing_data.csv",index=False)
```
